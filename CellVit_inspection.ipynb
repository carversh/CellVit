```python
p_anal_meta = 0.3367
p_anal_EUR = 0.3499
p_gist_meta = 0.2601
p_gist_EUR = 0.2623

OR_anal_meta = 1.601790
OR_anal_PROFILE_EUR = 2.073630
OR_GIST_meta = 1.958230	
OR_GIST_PROFILE_EUR = 2.928010


# define function that computes PAR
def compute_PAR(p, OR):
    """
    Compute Population Attributable Risk (PAR) given prevalence (p) and odds ratio (OR).
    
    Parameters:
    p (float): Prevalence of the risk factor in the population.
    OR (float): Odds Ratio associated with the risk factor.
    
    Returns:
    float: Population Attributable Risk (PAR).
    """
    return ( (p * (OR - 1)) / ((p * (OR -1)) + 1) )

anal_meta_ALL = compute_PAR(p_anal_meta, OR_anal_meta)
anal_PROFILE_EUR = compute_PAR(p_anal_EUR, OR_anal_PROFILE_EUR) 
GIST_meta_ALL = compute_PAR(p_gist_meta, OR_GIST_meta)
GIST_PROFILE_EUR = compute_PAR(p_gist_EUR, OR_GIST_PROFILE_EUR)

print(f'Anal Cancer Meta: {anal_meta_ALL:.4f}')
print(f'Anal Cancer PROFILE EUR: {anal_PROFILE_EUR:.4f}')
print(f'GIST Meta: {GIST_meta_ALL:.4f}')
print(f'GIST PROFILE EUR: {GIST_PROFILE_EUR:.4f}')
```

```python
import json
import numpy as np
import pandas as pd
from skimage.draw import polygon
from skimage.measure import label, regionprops

# Load the cells.json
with open("/Users/shayecarver/Downloads/test_results_graph/BL-18-T33931/BL-18-T33931/cells.json") as f:
    data = json.load(f)

cells = data['cells']  # list of cell entries
type_map = data['type_map']  # e.g., {'1': 'Neoplastic', ...}

# Convert type_map to int keys
type_map = {int(k): v for k, v in type_map.items()}

# Collect features
features = []
```

```python
for idx, cell in enumerate(cells):
    contour = cell['contour']  # list of [y, x] coordinates

    # Convert to np arrays
    y_coords, x_coords = zip(*contour)
    y_coords = np.array(y_coords)
    x_coords = np.array(x_coords)

    # Get bounding box size
    min_y, max_y = y_coords.min(), y_coords.max()
    min_x, max_x = x_coords.min(), x_coords.max()
    height = max_y - min_y + 1
    width = max_x - min_x + 1

    # Shift coords into mask space
    rr, cc = polygon(y_coords - min_y, x_coords - min_x, (height, width))
    mask = np.zeros((height, width), dtype=np.uint8)
    mask[rr, cc] = 1

    # Get shape features
    props = regionprops(label(mask))[0]

    features.append({
        "cell_id": idx,
        "type_label": type_map.get(cell['type'], 'Unknown'),
        "area": props.area,
        "perimeter": props.perimeter,
        "eccentricity": props.eccentricity,
        "solidity": props.solidity,
        "orientation": props.orientation,
        "major_axis_length": props.major_axis_length,
        "minor_axis_length": props.minor_axis_length,
        "aspect_ratio": props.major_axis_length / props.minor_axis_length
                        if props.minor_axis_length != 0 else 0,
        "circularity": 4 * np.pi * props.area / (props.perimeter**2)
                        if props.perimeter != 0 else 0,
        "centroid_x": cell["centroid"][1],
        "centroid_y": cell["centroid"][0],
    })

# Convert to DataFrame
df_features = pd.DataFrame(features)
```

```python
import json
import os
import pandas as pd
from tqdm import tqdm

# Path to the directory containing the folders
base_path = "Downloads/test_dir"

# List to store summary information per slide
summary_data = []

# Iterate through each item in the directory
for folder_name in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder_name)
    json_path = os.path.join(folder_path, "cell_detection.json")

    if not os.path.isdir(folder_path) or not os.path.exists(json_path):
        continue  # Skip if not a directory or cells.json missing

    print(f'Processing {folder_name}')

    with open(json_path) as f:
        data = json.load(f)

    type_map = {int(k): v for k, v in data.get('type_map', {}).items()}
    cells = data.get('cells', [])

    # Count cells by type
    counts = {v: 0 for v in type_map.values()}
    for cell in cells:
        cell_type = cell.get('type')
        type_label = type_map.get(cell_type, 'Unknown')
        counts[type_label] = counts.get(type_label, 0) + 1

    # Add to summary data
    summary_data.append({
        "slide_id": folder_name,
        **counts
    })

# Convert to DataFrame and save
df_summary = pd.DataFrame(summary_data)
df_summary.to_csv("colorectal_ocelot_cell_counts.csv", index=False)
print("Saved cell counts summary to colorectal_ocelot_cell_counts.csv")
```
